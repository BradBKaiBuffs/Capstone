**Q1 (1 point):Load data from ```socialmediapostdata.csv``` file using Pandas**

````python
# put your code here
# Read CSV
post_data = pd.read_csv('socialmediapostdata.csv')

# Show
post_data


````

**Q2 (1 point): Clean up the dataset and remove these colution```['country', 'id', 'language', 'latitude', 'longitude']```**

````python
# put your code here
# Clean up the dataset and remove columns
post_data = post_data.drop(columns=['country', 'id', 'language', 'latitude', 'longitude'])

# Show
post_data


````

````python
# sample output
````

**Q3  (1 point):  Calculate daily number of posts created by each user**

````python
# put your code here

# Convert 'date_time'
post_data['date_time'] = pd.to_datetime(post_data['date_time'], format='%d/%m/%Y %H:%M').dt.date

# Group by 'author' and 'date_time', then count the number of contents for each group
content_counts = post_data.groupby(['author', 'date_time'])['content'].count().reset_index(name='content_count')

# Show
content_counts
````

````python
# sample output
````

**Q4 (3 points):: Plot the number of posts per author per day calculated above using ```matplotlib.pyplot``` or ```plotly``` package.**

````python
# put your code here
# Convert to datetime format
content_counts['date_time'] = pd.to_datetime(content_counts['date_time'])

# Create graph
fig = px.line(content_counts, x='date_time', y='content_count', color='author', markers=True)

# Add title, labels
fig.update_layout(title='Number of Posts per Author per Day', xaxis_title='Date', yaxis_title='Number of Posts')

# Show
fig.show()




````

````python
## sample output
````

**Q5 (4 points): Create a chart to show the number of daily posts of author ```jimmyfallon```**

````python
# put your code here
# Filter the df
jimmy_fallon_data = content_counts[content_counts['author'] == 'jimmyfallon'].copy()

# Format date for plot
jimmy_fallon_data['date_time'] = pd.to_datetime(jimmy_fallon_data['date_time'])

# Plot graph
fig = px.line(jimmy_fallon_data, x='date_time', y='content_count', title='Number of Posts by Jimmy Fallon per Day')

# Add labels
fig.update_layout(xaxis_title='Date', yaxis_title='Number of Posts')

# Show
fig.show()
````

**Q6 (2 points): Extract post contents of author ```jimmyfallon```**

````python
# put your code here
# Extract posts created by 'jimmyfallon'
jimmy_fallon_posts = post_data[post_data['author'] == 'jimmyfallon']

# Pull 'content' column
jimmy_fallon_content = jimmy_fallon_posts['content']

# Show
jimmy_fallon_content


````

````python
# sample output
````

**Q7 (4 points):: Create a Word Cloud for posts of ```jimmyfallon```**

````python
# put your code here
# Import WordCloud and STOPWORDS
from wordcloud import WordCloud, STOPWORDS

# Combine all post contents
all_content = ' '.join(jimmy_fallon_content.astype(str))

# Add customized stopwords
stop_words = STOPWORDS.update(["https", "co", "t"])

# Create a WordCloud
wordcloud = WordCloud(stopwords = stop_words, width=800, height=400, background_color='white').generate(all_content)

# Set up matplotlib figure
plt.figure(figsize=(10, 5))

# Show the word cloud image
plt.imshow(wordcloud)

 # Hide the x and y axes
plt.axis('off')

# Show
plt.show()


````

**Q8 (4 Points): Plot number of daily likes and daily shares of `jimmyfallon`**

````python
# put your code below
# Rename columns used
jimmy_fallon_posts_renamed = jimmy_fallon_posts.rename(columns={'number_of_likes': 'daily_likes', 'number_of_shares': 'daily_shares'})

# Group data by date and get daily likes and shares
jimmy_fallon_posts_grouped = jimmy_fallon_posts_renamed.groupby('date_time')[['daily_likes', 'daily_shares']].sum().reset_index()

# Melt df
df_melt = pd.melt(jimmy_fallon_posts_grouped, id_vars=['date_time'], value_vars=['daily_likes', 'daily_shares'], var_name='Metrics', value_name='Count')

# Create plot
fig = px.line(df_melt, x='date_time', y='Count', color='Metrics', title='Daily Likes and Shares for Jimmy Fallon')

# Add x and y axis labels
fig.update_layout(xaxis_title='Date', yaxis_title='Count')

# Show
fig.show()




````
