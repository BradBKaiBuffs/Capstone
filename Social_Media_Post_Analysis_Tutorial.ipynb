### **1.Loading data using Pandas**

````python
#import pandas as pd
import pandas as pd

# uses the pandas package (referred to as pd) to read a CSV file named 'socialmediapostdata.csv' into a DataFrame called post_data.
# The encoding='utf-8' part specifies that the file uses UTF-8 encoding, which is a standard text encoding for representing a wide array of characters (like emoji)
post_data = pd.read_csv('socialmediapostdata.csv', encoding='utf-8')
````

````python
# show details of post_data
post_data
````

### **2.Clean and Visualizating Data**

In the following code cells, we will clean up our dataset by removing unnecessary columns. Then, we will visualize how often each author posts.

````python
# remove unnecessary columns including country, id, language, latitude, longitude
# The axis=1 parameter specifies that columns will be removed (to delete rows, it should be axis=0).
post_data = post_data.drop(['country', 'id', 'language', 'latitude', 'longitude'], axis=1)
post_data
````

Then, we will calculate the daily number of posts created by each user.

````python
# Convert 'date_time' to datetime type
# The original code assumes the format is MM/DD/YYYY, but it's actually DD/MM/YYYY
# We need to specify the correct format using the 'format' argument
post_data['date_time'] = pd.to_datetime(post_data['date_time'], format='%d/%m/%Y %H:%M').dt.date

# Group by 'author' and 'date_time', then count the number of contents for each group
content_counts = post_data.groupby(['author', 'date_time'])['content'].count().reset_index(name='content_count')

# show count data
content_counts
````

Next, visualize the ```content_counts``` calculated above using the visualization package ```matplotlib.pyplot```.

````python
# Import the matplotlib package's pyplot module for data visualization
import matplotlib.pyplot as plt

# Convert the 'date_time' column to datetime format for accurate plotting
content_counts['date_time'] = pd.to_datetime(content_counts['date_time'])

# Prepare to plot the data, initialize a new figure with a specified size (width=12 inches, height=6 inches)
plt.figure(figsize=(12, 6))

# Loop through each unique author to plot their post frequencies
for author in content_counts['author'].unique():

    # Filter data for the current author
    author_data = content_counts[content_counts['author'] == author]

    # Plot the number of posts for each author over time, using circles ('o') as markers
    plt.plot(author_data['date_time'], author_data['content_count'], marker='o', label=author)

# Set the title of the plotting
plt.title('Number of Posts per Author per Day')
# Label the x-axis as 'Date'
plt.xlabel('Date')
# Label the y-axis as 'Number of Posts'
plt.ylabel('Number of Posts')
# Display a legend to identify each author's data series, with 'Author' as the legend title
plt.legend(title='Author')
# Automatically adjust subplot parameters to give specified padding and ensure everything fits without overlap
plt.tight_layout()
# Display the plotting
plt.show()
````

In the last code cell, we plotted post counts for all authors. How can we plot the data for a specific author? Let's plot data for the author ```taylorswift13``` as an example.

````python
# Filter the DataFrame for 'taylorswift13' only
taylor_swift_data = content_counts[content_counts['author'] == 'taylorswift13'].copy()

# Ensure that 'date_time' is in datetime format for plotting
taylor_swift_data['date_time'] = pd.to_datetime(taylor_swift_data['date_time'])

# Plot
plt.figure(figsize=(10, 5))  # Set the figure size
plt.plot(taylor_swift_data['date_time'], taylor_swift_data['content_count'], linestyle='-', color='blue')

plt.title('Number of Posts by Taylor Swift per Day')
plt.xlabel('Date')
plt.ylabel('Number of Posts')
plt.grid(axis='y', linestyle='--')  # Add horizontal grid lines for easier reading
plt.tight_layout()  # Adjust layout to not cut off labels
plt.show()
````

### **3.Create Interactive Plotting**

So far, we have learned how to extract specific data and represent it in static figures. However, there are times when we need our visualizations to be more interactive. To address this, we will introduce another powerful visualization tool called [Plotly](https://plotly.com/)

Plotly is an interactive, open-source plotting package that supports a wide range of charts and visualizations. It's particularly known for its ability to produce complex, interactive plots that can be embedded in web applications or displayed in Jupyter notebooks and Python scripts. Plotly supports multiple programming languages, including Python, R, and JavaScript, making it versatile for data science, engineering, and analytics tasks.


[Plotly Python package](https://plotly.com/python/)



````python
# Import the Plotly Express module for easy-to-use data visualization functions
import plotly.express as px

# Convert the 'date_time' column to datetime format to ensure accurate plotting of time series data
content_counts['date_time'] = pd.to_datetime(content_counts['date_time'])

# Use Plotly Express to create a line plot.
fig = px.line(
              content_counts,  # The DataFrame containing the data to plot
              x='date_time',  # Specify the DataFrame column to use for the x-axis values
              y='content_count',  # Specify the DataFrame column to use for the y-axis values
              color='author',  # Differentiate data lines by 'author' column, assigning each author a unique color
              markers=True,  # Add markers to each data point for better visibility
              labels={'content_count': 'Number of Posts', 'date_time': 'Date'},  # Rename axis labels for clarity
              title='Number of Posts per Author per Day')  # Set the title of the plot

 # Display the interactive plot.
fig.show()
````

````python
# Create an interactive chart for data of Taylor Swift using Plotly python package
fig = px.line(taylor_swift_data, x='date_time', y='content_count',
             title='Number of Posts by Taylor Swift per Day',
             labels={'date_time': 'Date', 'content_count': 'Number of Posts'},
             markers=True)

# Show the figure
fig.show()
````

### **4.Create a Word Cloud for Posts**

A word cloud is a visual representation of text data where the size of each word indicates its frequency or importance in the text. It's a popular tool for text analysis, especially for highlighting popular or significant terms in large text datasets, such as speeches or social media posts.

In a word cloud, commonly used words (also called stop words, e.g.: the, and, or) are typically omitted to focus on the most relevant words. This visualization technique helps in quickly perceiving the most prominent terms, making it useful for data exploration, presentation, and qualitative analysis.

````python
post_data
````

````python
# extract posts created by author 'taylorswift13'
taylor_swift_posts = post_data[post_data['author'] == 'taylorswift13']

# Select only the 'content' column for these rows
taylor_swift_content = taylor_swift_posts['content']
taylor_swift_content
````

````python
# Import the WordCloud, and STOPWORDS tools from the wordcloud package
from wordcloud import WordCloud, STOPWORDS

# Combine all post contents into a single string, each post content is converted to string datatype and separated by spaces
all_content = ' '.join(taylor_swift_content.astype(str))

# wordcloud package has a set of stopwords. We can add customized stopwords with update methods to remove them from text data
updated_stop_words = STOPWORDS.update(["https", "co", "t"])

# Create a WordCloud with specific dimensions and background color, then generate a word cloud image from the combined text
wordcloud = WordCloud(stopwords = updated_stop_words, width=800, height=400, background_color='white').generate(all_content)

# Set up a matplotlib figure with a specific figure size to display the word cloud
plt.figure(figsize=(10, 5))

# Show the word cloud image in the figure.
plt.imshow(wordcloud)

 # Hide the x and y axes to focus solely on the word cloud image
plt.axis('off')

 # Render and display the figure containing the word cloud
plt.show()
````
